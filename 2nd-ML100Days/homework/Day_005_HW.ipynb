首先，在去除掉CODE_GENDER中被記載為XNA的遺漏值之後，我們可以關查性別與AMT_CREDIT的關係

```python
import seaborn as sns
app_train = app_train.loc[app_train['CODE_GENDER']!='XNA',]
sns.boxplot(x='CODE_GENDER',y='AMT_CREDIT',data=app_train)
```

由上圖，我們並不能簡單的看出不同性別間的AMT_CREDIT是否有不同<br>
改觀察性別與是否擁有汽車後，發現男性擁有汽車的比沒有的還多，但女性擁有汽車的遠少於沒有的。

```python
sns.countplot(x = 'FLAG_OWN_CAR',hue = 'CODE_GENDER',data = app_train)
```
