- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
plt.figure()
fignum = 1

for i in np.arange(1,4):
    for j in np.arange(0,i):
        plt.subplot(2,3,fignum)
        plt.scatter(X[:,i],X[:,j],c=y)
        fignum = fignum+1
        
plt.show
```

```python
Model_n2 = KMeans(n_clusters=2)
Model_n2.fit(X)
plt.figure()
fignum = 1

for i in np.arange(1,4):
    for j in np.arange(0,i):
        plt.subplot(2,3,fignum)
        plt.scatter(X[:,i],X[:,j],c=Model_n2.labels_.astype(float))
        fignum = fignum+1
        
plt.show
```

```python
Model_n3 = KMeans(n_clusters=3)
Model_n3.fit(X)
plt.figure()
fignum = 1

for i in np.arange(1,4):
    for j in np.arange(0,i):
        plt.subplot(2,3,fignum)
        plt.scatter(X[:,i],X[:,j],c=Model_n3.labels_.astype(float))
        fignum = fignum+1
        
plt.show
```

```python
Model_n3_init20 = KMeans(n_clusters=3,n_init=20)
Model_n3_init20.fit(X)
plt.figure()
fignum = 1

for i in np.arange(1,4):
    for j in np.arange(0,i):
        plt.subplot(2,3,fignum)
        plt.scatter(X[:,i],X[:,j],c=Model_n3_init20.labels_.astype(float))
        fignum = fignum+1
        
plt.show
```

```python
STD = StandardScaler()
STD.fit(X)
X = STD.transform(X)
Model_n3_std = KMeans(n_clusters=3)
Model_n3_std.fit(X)
plt.figure()
fignum = 1

for i in np.arange(1,4):
    for j in np.arange(0,i):
        plt.subplot(2,3,fignum)
        plt.scatter(X[:,i],X[:,j],c=Model_n3_std.labels_.astype(float))
        fignum = fignum+1
        
plt.show
```

```python
from sklearn.metrics import accuracy_score
for i in range(len(Model_n3_std.labels_)):
    if(Model_n3_std.labels_[i]==1):
        Model_n3_std.labels_[i] = 0
    elif(Model_n3_std.labels_[i]==0):
        Model_n3_std.labels_[i] = 1
        
print('Accuracy:%.4f' %(accuracy_score(y,Model_n3_std.labels_)))
```

```python
for i in range(len(Model_n3_init20.labels_)):
    if(Model_n3_init20.labels_[i]==1):
        Model_n3_init20.labels_[i] = 0
    elif(Model_n3_init20.labels_[i]==0):
        Model_n3_init20.labels_[i] = 1
        
print('Accuracy:%.4f' %(accuracy_score(y,Model_n3_init20.labels_)))
```

```python
Model_n3_std_init20 = KMeans(n_clusters=3,n_init=20)
Model_n3_std_init20.fit(X)
plt.figure()
fignum = 1

for i in np.arange(1,4):
    for j in np.arange(0,i):
        plt.subplot(2,3,fignum)
        plt.scatter(X[:,i],X[:,j],c=Model_n3_std_init20.labels_.astype(float))
        fignum = fignum+1
        
plt.show
```

```python
for i in range(len(Model_n3_std_init20.labels_)):
    if(Model_n3_std_init20.labels_[i]==1):
        Model_n3_std_init20.labels_[i] = 0
    elif(Model_n3_std_init20.labels_[i]==0):
        Model_n3_std_init20.labels_[i] = 1
        
print('Accuracy:%.4f' %(accuracy_score(y,Model_n3_std_init20.labels_)))
```
