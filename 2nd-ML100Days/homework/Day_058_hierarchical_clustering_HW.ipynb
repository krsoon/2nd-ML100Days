```python
# 載入套件
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
```

```python
# 設定 2D 樣板資料
n_samples = 1500
random_state = 100

# 生成 雙半月 資料點
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)

# 生成 緊密三群 資料點
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)

# 生成 2維均勻分布 資料點
no_structure = np.random.rand(n_samples, 2), None
```

```python
# 設定群集與繪圖參數
plt.figure(figsize=(48, 6))
plot_num = 1
datasets = [('noisy_moons',noisy_moons, 2), ('blobs',blobs, 3), ('no_structure',no_structure,4)]
```

# 作業
* 試著完成剩下三種 2D 樣板資料的繪圖, 並觀察有何不同

```python
# 執行各種樣板資料的繪圖迴圈
from sklearn.cluster import AgglomerativeClustering
estimators = ['ward','complete','average']

for name,data,n_clusters in datasets:
    X = data[0]
    Y = data[1]
    for linkage in estimators:
        model = AgglomerativeClustering(n_clusters=n_clusters,linkage=linkage)
        model.fit(X)
        labels = model.labels_
        fig = plt.figure(plot_num, figsize=(4, 4))
        plt.scatter(X[:,0],X[:,1],c = labels)
        plt.title('%s n_clusters:%d linkage:%s' %(name,n_clusters,linkage))
        plot_num = plot_num+1
    fig = plt.figure(plot_num, figsize=(4, 4))
    plt.scatter(X[:,0],X[:,1],c = Y)
    plt.title('Ground Truth')
    plot_num = plot_num+1

```
